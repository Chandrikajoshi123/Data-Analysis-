```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(r'\s*\(.*\)', '', regex=True)  # I have done this to remove any parenthesis
      .str.replace(' ', '_')
)
```

```python
#percentage coloumn seems unnecessary as it has only 1 unique value
df.drop('percentage',axis=1, inplace=True)
```

```python
df.head()
```

```python
df['completion_rate_clean'] = df['completion_rate'].str.extract(r'^(\d+\.\d+)')
```

```python
df.drop('completion_rate',axis=1, inplace=True)
df.head()
```

```python
#cleaning the rating to be out of 5 
def fix_rating(val):
    val = str(val).replace('.', '')  
    if len(val) >= 2:
        return float(val[0] + '.' + val[1])
    elif len(val) == 1:
        return float(val + '.0')
    else:
        return None

df['rating_clean'] = df['rating'].apply(fix_rating)
df.head()
```

```python
#sine i have rating_clean already no need for rating coloumn
df.drop('rating',axis=1, inplace=True)
df.head()
```

```python
#converting completion_rate_clean into float
df['completion_rate_clean'] = df['completion_rate_clean'].astype(float)

print(df.dtypes)
```

```python
df.describe()
#completion rate has max value 999.872000 which seems to be an issue as it should be out of 100 
```

```python
df['completion_rate_clean'].head(10)
```

```python
df = df[df['completion_rate_clean'] <= 100]
df['completion_rate_clean'].describe()
#scaled the completion rate to be out of 100
```

```python
df.to_csv('cleaned_course_data.csv', index=False)
```
